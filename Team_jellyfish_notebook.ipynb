Despite the crucial role played by higher education institutions in providing knowledge and skills, student dropout and academic success continue to be significant challenges worldwide. Therefore, understanding the factors that contribute to student dropout and academic success is crucial for educational institutions, policymakers, and educators.

The dataset also includes information about the courses chosen by the students, their application mode, marital status, and other relevant information available at the time of enrollment.

By analyzing this dataset, we can identify the predictors of student dropout and academic success across a wide range of disciplines offered at a higher education institution. This analysis can provide valuable insights into what motivates students to stay in school or abandon their studies, leading to effective interventions that support student retention and success.

## Comparing the academic performance of scholarship holders and non-scholarship holders, considering the course taken by the students.

```python
df["Scholarship holder"].unique() # inspecting to see if only 0 and 1.
pd.set_option('display.max_columns', None)
```

```python
scholarships = df[df["Scholarship holder"] == 1]

# dictionary to remap "Course" column
rename_courses = {1 : "Biofuel", 2 : "Animation", 3: "Social Service(evening)", 4: 'Agronomy'\
                  , 5: "Communcations", 6: "Veterinary Nursing", 7:"Informatics", 8: "Equiniculture", 9: "Management"\
                  ,10: "Social Services", 11: "Tourism", 12: "Nursing", 13: "Oral Hygiene", 14: "Marketing",\
                  15: "Journalism",16: "Basic Education", 17: "Manaegement (evening)"}

scholarships = scholarships.replace({"Course": rename_courses})

non_scholarships = df[df["Scholarship holder"] == 0]

non_scholarships = non_scholarships.replace({"Course": rename_courses})

print(scholarships["Course"].head()) # doubling checking change has been made in the column
print(non_scholarships["Course"].head())
```

```python
table = pd.pivot_table(scholarships, 
               values= 'Gender',
               index=['Target'],
               columns=['Course'],
               aggfunc='count'
              )

table2 = pd.pivot_table(non_scholarships, 
               values= 'Gender',
               index=['Target'],
               columns=['Course'],
               aggfunc='count'
              )
ax1 = table.T.plot(kind='bar', ylabel='Count', figsize=(5, 5), title = "Students with Scholarships", xlabel = "")
ax1.legend(loc="upper left")
ax2 = table2.T.plot(kind='bar', ylabel='Count', figsize=(5, 5), title = "Students with no Scholarships", xlabel = "")
ax2.legend(loc="upper left")
```

## Observed:
- Students with scholarships have better overall performance than those with no scholarships.
- Those with scholarships have significantly lower drop out rates across all courses, especially in evening management courses, veterinary nursing, agronomy and marketing degrees.
- Only top students usually receive scholarships, and the difference in performance clearly indicates this.
- No students received any scholarships in the Biofuel major/degree

## 6.     Investigate the effect of previous qualifications on student performance in higher education, considering other factors such as age, gender, and course.

```python
# numbers of previosuly qualified vs previously unqualified higher education
qualified = [2, 3, 4, 5, 15, 16, 17]
not_qualified = [1, 6, 7, 8, 9, 10, 11, 12, 13, 14]

# creating a dataframe with all higher educated individuals
higher_education = (df["Previous qualification"] == 2) | (df["Previous qualification"] == 3)\
| (df["Previous qualification"] == 4)| (df["Previous qualification"] == 5) | (df["Previous qualification"] == 15)\
| (df["Previous qualification"] == 16) | (df["Previous qualification"] == 17)

previous_higher_education = df[higher_education]

rename_courses = {1 : "Biofuel", 2 : "Animation", 3: "Social Service(evening)", 4: 'Agronomy'\
                  , 5: "Communcations", 6: "Veterinary Nursing", 7:"Informatics", 8: "Equiniculture", 9: "Management"\
                  ,10: "Social Services", 11: "Tourism", 12: "Nursing", 13: "Oral Hygiene", 14: "Marketing",\
                  15: "Journalism",16: "Basic Education", 17: "Manaegement (evening)"}

rename_Genders = {0: "Female", 1: "Male"}

previous_higher_education = previous_higher_education.replace({"Course": rename_courses, "Gender": rename_Genders})

# creating a dataframe with secondary educated individuals
only_secondary_education = df[~higher_education]
```

```python
# testing all data included

(df.shape[0]) == previous_higher_education.shape[0] + only_secondary_education.shape[0]
```

```python
# testing for correct number for previously highly educated individuals
for i in qualified:
    if i in previous_higher_education["Previous qualification"].unique():
        print("Included")
    else:
        print("Excluded", i)
# all good
```

```python
for i in not_qualified:
    if i in only_secondary_education["Previous qualification"].unique():
        print("Included")
    else:
        print("Excluded", i)
# all good
```

So there is no clear indication that previous education improves the likelihood of not dropping out since individuals situations differ.
